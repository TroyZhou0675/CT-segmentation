```python
from metrics import multiclass_dice as dice
from metrics import multiclss_soft_iou as iou
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import os
import tensorflow as tf
```

```python
def show_img(img):
    plt.imshow(img)
    plt.show()
```

```python
img_path = 'D:/0-MyDoc/DeepLearning/MIS/CT_SEG/data/processed/masks/'
Number_of_Class = 4

Pic_name = os.listdir(img_path)
img = cv.imread(img_path+Pic_name[0],cv.IMREAD_GRAYSCALE)
img2 = cv.imread(img_path+Pic_name[1],cv.IMREAD_GRAYSCALE)

img_oh = np.eye(4,dtype = np.uint8)[img]
img_oh_tf = tf.convert_to_tensor(img_oh)
img_oh_tf = tf.expand_dims(img_oh_tf,axis = 0)

img2_oh = np.eye(4,dtype = np.uint8)[img2]
img2_oh_tf = tf.convert_to_tensor(img2_oh)
img2_oh_tf = tf.expand_dims(img2_oh_tf,axis = 0)

#print(img_oh_tf.shape)
x = dice(img_oh_tf,img2_oh_tf,ignore_background=True) 
print(f"accuracy is {x.numpy().item()}")
show_img(img)
show_img(img2)
```
